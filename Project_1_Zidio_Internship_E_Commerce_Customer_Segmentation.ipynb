```python
x= df.iloc[:,2:].values
x
```

```python
scale=MinMaxScaler()
features=scale.fit_transform(x)
features
```

```python
inertia=[]
for i in range(1,10):
  k_means=KMeans(n_clusters=i)
  k_means.fit(features)
  inertia.append(k_means.inertia_)
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1, 2, 1)
plt.plot(range(1, 16), inertia, 'bo-')
plt.xlabel("No of clusters")
plt.ylabel('Inertia')
plt.show()
```

```python
plt.subplot(1,2,2)
kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k=(1, 15))
visualizer.fit(features)
plt.title("Elbow Graph using Elbow Visualizer")
visualizer.show()
plt.show()
```

```python
silhouette_avg = []
for i in range(2, 10):
    # Initialize KMeans cluster
    kmeans = KMeans(n_clusters=i)
    cluster_labels = kmeans.fit_predict(features)
    # Calculate silhouette score
    silhouette_avg.append(silhouette_score(features, cluster_labels))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(2,10), silhouette_avg[:8], 'bX-')  # Adjust the range and slicing
plt.xlabel('Values of K')
plt.ylabel('Silhouette score')
plt.title('Silhouette analysis for optimal K')
plt.show()
```

```python
model = KMeans(n_clusters=3)
model=model.fit(features)
```

```python
y_km=model.predict(features)
centers=model.cluster_centers_
```
